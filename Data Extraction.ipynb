### Here, I have scraped website articles (which is about AI) using  Beautiful Soup. 
#### To scrap the Headings & Contents of articles and save it to a text file.

```python
#!pip install requests
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
# providing a user agent to the request. By doing so, the website will think that someone is actually visiting the site using a web browser.
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:60.0) Gecko/20100101 Firefox/60.0"
}
```

```python
source= requests.get('https://insights.blackcoffer.com/ai-in-healthcare-to-improve-patient-outcomes/', headers= headers)
```

```python
# creating a soup object to extract text data from site, parser used is html parser
soup= BeautifulSoup(source.text, 'html.parser')
```

```python
print(soup)
```

```python
# In an HTML page tags contains blocks of texts, 
# click on topics of inside webpage -> select inspect , we get code corresponding to each block of text.
```

Find(): This method finds the first matched element.
Find_all(): This method finds all the matched elements.

```python
# bUt we'll try to access all the only <p> tags, since it contains only texts
# So, <p> tags contain two things : Title of paragraph and Paragraph
content= soup.find('div', class_= 'td-post-content').find_all('p')
print(content)
```

```python
# to get all the texts paragraphs
l= []
for heading in content:
    l.append(heading.text)
```

```python
print(l)
```

```python
# saving list of text data in a text file, creating file first
with open('C:/Users/hp/Machine Learning/scrape4.txt','w') as file:
    file.write(str(l))
```

### Scraping multiple URLs at once


```python
url = ['https://insights.blackcoffer.com/ai-in-healthcare-to-improve-patient-outcomes/', 'https://insights.blackcoffer.com/what-if-the-creation-is-taking-over-the-creator/', 'https://insights.blackcoffer.com/what-jobs-will-robots-take-from-humans-in-the-future/']


for pg in url:
    source = requests.get(pg, headers= headers)
    soup = BeautifulSoup(source.text, "html.parser")
    content= soup.find('div', class_= 'td-post-content').find_all('p')
    print(content)
    
```

```python
l= []
for heading in content:
    l.append(heading.text)
```

```python
print(l)
```
